```python
# mengubah teks menjadi lower

data['cleaned_text'] = data['cleaned_text'].str.lower()
data.head()
```

```python
# normalisasi

df_alay = pd.read_csv('/content/spilkada_banten24.csv', encoding='ISO-8859-1', header=None)
df_alay = df_alay.rename(columns={0: 'alay', 1: 'formal'})
df_alay_dict = dict(zip(df_alay['alay'], df_alay['formal']))
```

```python
def normalize_text(text):
  words = text.split()
  normalized_words = [df_alay_dict.get[word] if word in df_alay_dict else word for word in words]
  return ' '.join(normalized_words)

data['cleaned_text'] = data['cleaned_text'].apply(normalize_text)
```
